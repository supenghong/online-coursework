# Question 1: Integers Mod Prime


## Modular arithmetic

This question explores some of the ideas surrounding what's known as ***modular arithmetic***. The arithmetic of a clock with different numbers of hours is the best way to think about this concept.

In the case of a 12-hour clock 

<h1><center> 9  +  4 = 1 </center></h1>

<img src="https://upload.wikimedia.org/wikipedia/commons/a/a4/Clock_group.svg" width=300/>

In a more mathematical lingo, we say that 

$$
\Large 9 + 4 \ = \ 13 \ = \ 1 + 12 \ = \ 1 \ \ \text{mod} \ \ 12.
$$

We can do multiplication in a similar way 

$$
\Large 7 \times  4 \ = \ 28 \ = \ 4 + 2 \times 12 \ = \ 4 \ \ \text{mod} \ \ 12.
$$

<h3><center> BASIC IDEA :  Modular arithmetic only cares about the remainder after dividing by given number p.  </center></h3>

### A problem with division 

Here is something that is weird about modular arithmetic if we are not careful:

$$
\Large 6 \times  4 \ = \ 24 \ = \ 2 \times 12 \ = \ 0 \ \ \text{mod} \ \ 12,
$$

or

$$
\Large 4 \times  3 \ = \ 12 \ = \  0 \ \ \text{mod} \ \ 12.
$$

Multiplying two positive number and getting 0 doesn't leave us with enough interesting mathematical structure to do very much.

**For example:** We can't ***divide*** numbers on the clock.

As is always the case with numbers, zero times anything gives zero.

$$
\Large \frac{0}{3} = 0 \ \ \text{mod} \ \ 12 \quad \iff \quad  0 \times 3  \ = \ 0  \ \ \text{mod} \ \ 12.
$$

However, 

$$
\Large \frac{0}{3} = 4 \ \ \text{mod} \ \ 12   \quad \iff \quad  4 \times 3  \ = \ 0  \ \ \text{mod} \ \ 12.
$$

And also

$$
\Large \frac{0}{3} = 8 \ \ \text{mod} \ \ 12 \quad \iff \quad  8 \times 3  \ = \ 0 \ \ \text{mod} \ \ 12.
$$

<h2><center> We would like division to only have one answer. </center> </h2>


## Integers mod p


The problem with division happened not because the product of two numbers equaled zero exactly. Problems happened because several products of two numbers equal exactly 12. But this would not be possible if 12 had no factors other than 1 and itself. That is another way of saying that everything would work if 12 was ***prime***, which it is definitely *not*. 

<h2><center> Everything works nicely if we consider arithmetic modulo a PRIME number. </center> </h2>

\Large \ \text{AND} \quad \quad \longleftrightarrow \quad \quad \times \quad \text{mod} \quad 2

But there isn't a correspondence between $\text{OR}$ and $+$. 


### Exclusive OR  =  XOR

I haven't mentioned it yet but $\text{AND}$, $\text{OR}$ are not the only binary operations that act on $\text{True}$ and $\text{False}$. There are actually a lot of operations that take two Boolean variables and give you back another Boolean variable. A key point is that *any* two of the many operators are enough to make any of the others. 

There is another very common binary operation called ***Exclusive OR***, or $\text{XOR}$. This operator has the table

$$
\Large
\begin{array}{cccccc}
 {\normalsize \text{XOR}} &|& 0 & 1  \\
 - &|& - & -  \\
 0 &|& 0 & 1  \\
 1 &|& 1 & 0  \\
\end{array}
$$

In words, $\text{XOR}$ basically means "One, or the other, but not both." And now the big reveal (if it wasn't already a dead giveaway):

$$
\Large \ \text{XOR} \quad \quad \longleftrightarrow \quad \quad + \quad \text{mod} \quad 2
$$

I already said that any two Boolean operations are enough to make any of the others,

$$
\Large    A \ \text{xor} \  B  \  = \   ( \  A \ \text{or}  \  B  \ ) \ \text{and} \ (\ \text{not} \ (\  A \  \text{and} \ B \ ) \ )
$$

In Python (and many other programming languages) the exclusive or is represented with a $\, \wedge$.

<br>


```python
A = np.array([[False,  True]])
B = np.array([[False],[True]])

A_xor_B = A ^ B
A_or_B  = A | B
A_and_B = A & B

print('A and B:')
print(A_and_B)
print('')
print('A xor B:')
print(A_xor_B)
print('')
print('A or B')
print(A_or_B)
print('')
print('A xor B == (A or B) and (not (A and B)):')
print(A_xor_B ==  A_or_B & ( ~ A_and_B ) )
```

#### Extended Euclidean Algorithm

The way to solve this problem is to use something called the **Extended Euclidean Algorithm**. It is similar to the GCD Algorithm that we discussed in class. You should look up online how the general case works, and how you can implement it.

```python
## DO NOT ERASE THIS CELL ##

def inverse(a,p):
    
    ## INPUT YOUR CODE BELOW THIS LINE ## 
    if isinstance(a,int) == False or isinstance(p,int) == False:
        raise ValueError ('Not all inputs are integers.')
        
    elif a == 0:
        raise ValueError ('The input for a cannot be 0.')
        
    elif is_prime(p) == False:
        raise ValueError ('%s is not a prime number' %p)
        
    else:
        def egcd(a,b):
            if a == 0:
                return (b, 0, 1)
            g, y, x = egcd(b % a, a)
            return (g, x - (b//a)*y, y)
        g,x,y = egcd(a,p)
        if g != 1:
            raise ValueError ("Inverse does not exist!")
        else:
            return x % p
        
```

$\large \text{If } n \ne 0$, 

$$
\Large \text{factorial}(n,p) \ = \ n\times \text{factorial}(n-1,p) \quad \text{mod} \quad p.
$$

For this problem, you should require that $p$ is prime; and have the function raise an error if not.




```python
## DO NOT ERASE THIS CELL ##

def factorial(n,p):
    
    ## INPUT YOUR CODE BELOW THIS LINE ##
    if is_prime(p) == False:
        raise ValueError ('%s is not a prime number' %p)
    elif n == 0:
        return 1
    else:
        return (n*factorial(n-1,p))%p
```

# Question 2: Random Recursion Relations  

## Two deterministic examples

### Case +1

In lecture, we studied the two following recurrence relations

$$
\Large x_{n+1} \ = \ 2\, x_{n} \ + \ x_{n-1}, \quad \quad \text{or} \quad \quad x_{n+1} \ = \  x_{n} \ + \ x_{n-1}.
$$

The first recursion shows up when computing the continued fraction representation of $\sqrt{2}$. The second recursion is famous Fibonacci sequence, and shows up when studying the golden ratio $(1 + \sqrt{5})/2$.

In both examples it's convenient to set  
$$
\Large x_{0} \ = \ x_{1} \ = \ 1
$$

In general, the questions we care about here don't depend  much on the exact details of the initial conditions.

More generally, we could use a parameter, $\alpha$ to study more general recursion relations, 

$$
\Large  x_{n+1} \ = \ 2\alpha \, x_{n} \ + \ x_{n-1} \quad \quad \implies \quad \quad \large   \left[
\begin{array}{c}
 x_{N+1} \\
 x_{N} \\
\end{array}
\right] \ = \  \left[
\begin{array}{cc}
 2 \alpha  & 1 \\
 1 & 0 \\
\end{array}
\right]^{N} \left[
\begin{array}{c}
 1 \\
 1 \\
\end{array}
\right].
$$

The second formula results from considering the problem as a *matrix* recursion. 

### Case -1

Alternatively we could consider a slightly modified recursion formula

$$
\Large  x_{n+1} \ = \ 2\alpha \, x_{n} \ - \ x_{n-1} \quad \quad \implies \quad \quad \large   \left[
\begin{array}{c}
 x_{N+1} \\
 x_{N} \\
\end{array}
\right] \ = \  \left[
\begin{array}{cc}
 2 \alpha  & -1 \\
 1 & 0 \\
\end{array}
\right]^{N} \left[
\begin{array}{c}
 1 \\
 1 \\
\end{array}
\right] .
$$


We will see that this behaves in interesting ways compared to the +1 case.  In both cases, assume 

$$
\Large \alpha > 0. 
$$


## The Lyapunov growth factor


We showed in class that the numbers in these kinds of recursion grow very quickly. In fact they grow ***exponentially fast***. If a sequence is growing exponentially, then the following quantity is a useful measure of how fast, 

$$
\Large \lambda(\alpha) \ = \ \lim_{n\to\infty} |x_{n}|^{1/n}.
$$

This is called the ***Lyapunov growth factor***. In either of the above cases, the determinant of the matrix is $\mp 1$. This means that the product of the eigenvalues is $\mp 1$. This means that one eigenvalue has an absolute value greater one, and the other has an absolute value less than one. This means that repeatedly multiplying by a matrix will lead to the growth of the largest eigenvalue and the decay of the other. 

The summary is that in both of the above cases, $\lambda(\alpha)$ is the same as absolute value of the largest eigenvalue of the matrix




### For Case +1

$$
\Large \lambda(\alpha,+1) \ \equiv \ \alpha \ + \ \sqrt{\alpha^{2} + 1} .
$$

### For Case -1

$$
\Large \lambda(\alpha,-1) \ \equiv \ \left|\, \alpha \ + \ \sqrt{\alpha^{2} - 1}\, \right|  .
$$


There are two more points before we get into the random situation. 

* First,

$$
\Large \lambda(\alpha, -1) \ < \ \lambda ( \alpha, +1) \quad \quad \text{for all } \quad \quad 0 \ < \ \alpha 
$$

* Second, 

$$
\Large \lambda(\alpha, -1)  \ = \ 1 \quad \quad \text{for} \quad \quad 0 \ < \ \alpha \ \le 1.
$$

<br>



## The Random $\large \pm$ Case.

Now we want to randomly mix the above two examples.

$$
\Large  x_{n+1} \ = \ 2\alpha \, x_{n} \ \pm \ x_{n-1} \quad \quad \text{where each }\ \pm \ \text{ shows up with probability} \quad   \frac{1}{2}. 
$$

We are still assuming 

$$
\Large x_{0} \ = \ x_{1} \ = \ 1.
$$

Now we can compute 

$$
\Large \lambda(\alpha,\pm) \ = \ \lim_{n\to \infty} | x_{n} | ^{1/n}.
$$

Here is a critical fact about this quantity:
<h3><center> The Lyapunov factor will be the same (with probability=1) for all random +/- sequences.   </center></h3>
There is something deterministic in our random process.

You might think the random $\pm$ case might be simple to solve analytically. Half the time it's the $+1$, and the other half it's the $-1$. But it's not this simple. The reason is that the $\pm$ matrices don't commute. If they did, we could just re-order the sequence and compute the answer. But we can't. The only thing we can say for sure it that the solution to the random sequence satisfies the inequality 

$$
\Large \lambda(\alpha,-1) \ < \ \lambda(\alpha,\pm) \ < \ \lambda(\alpha,+).
$$



<br>

## A major hint.

If you compute numbers directly, you will find things get very big very fast. This will become a big problem for finite  precision very quick. The big trick in this case is to use Logarithms.

First, for *any* sequence, $x_{n}$ with $x_{0} \ne 0$, we can define the ratios

$$
\Large r_{n} \ = \ \frac{|x_{n}|}{|x_{n-1}|}  \quad \quad \text{for} \quad \quad n \ \ge \ 1.
$$

Now note the following simple algebraic fact 

$$
\Large |x_{n}|^{1/n} \ = \ |x_{0}|^{1/n}\,\exp \left[ \frac{1}{n} \sum_{k=1}^{n} \log( r_{k} ) \right].
$$

In the random sequence, you will have to be careful about when $x_{n}$ changes sign. But other than that, you should be able to use Logarithms to make you calculation much easier.

Write a function called `Lyapunov` that takes a single `float` argument, can computes 

$$
\Large \lambda(\alpha,\pm) .
$$

Your function should compute this to as many digits as you can within a runtime of less than 2-3 seconds. 

For a given $\alpha$, you will get most of the possible marks for computing the correct answer to 3 digit accuracy. Beyond 3 digits, you will be awarded more marks for computing more digits in under a few seconds. 

I will award special marks if you find a way to compute the answer to 10 or more accurate digits. 

**Note:** There are valid ways to solve this problem that involve randomly generating $\pm$. And there are valid ways to solve this problem that don't involve computing random numbers. Both ways can work well.

```python
## DO NOT ERASE THIS CELL ##

def Lyapunov(alpha):
    
    ## INPUT YOUR CODE BELOW THIS LINE ##
    if alpha <= 0:
        raise ValueError ("Input for alpha must be greater than zero.")
    
    n = int(1.2e5)
    sum = 0
    x = [1,1]
    
    for k in range(2,n):
        x[0], x[1] = x[1], 2*alpha*x[1] + (np.random.randint(0,2)*2-1)*x[0]
        
        if np.abs(x[1]) == float('inf'):
            return np.exp(sum)
        
        sum += 1/n * np.log(np.abs(x[1]/x[0]))
    
    return np.exp(sum)
```

You'll have to work out the $\alpha > 1$ case yourself.
